#### Capture Regions on Board

Given a 2-D board A of size N x M containing 'X' and 'O', capture all regions surrounded by 'X'. A region is captured by flipping all 'O's into 'X's in that surrounded region.

```python
def capture_region(mat):
    
    N = len(mat)
    M = len(mat[0])
    
    #queue to store nodes during BFS
    q = []
        
    for i in range(M):
        for j in range(N):
            if mat[i][j] == 'X':
                q.append((i,j))
    
```

```python
def main():
    A = [
           [X, X, X, X],
           [X, O, O, X],
           [X, X, O, X],
           [X, O, X, X] 
        ]
    print(capture_region(A))
if __name__ == '__main__':
    main()
```
