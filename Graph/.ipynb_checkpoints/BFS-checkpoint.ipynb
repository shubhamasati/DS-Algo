#### Black Shapes

Given character matrix A of O's and X's, where O = white, X = black. Return the number of black shapes. A black shape consists of one or more adjacent X's (diagonals not included)

```python
def is_safe(A, i, j, visited):
    N = len(A)
    M = len(A[0])
    
    if -1 < i < N and -1 < j < M and not visited[i][j] and A[i][j] == 'X':
        return True
    return False
```

```python
def BFS(self, grid, i, j, visited):
        
        movex = [-1, 0, 0, 1]
        movey = [ 0, -1, 1, 0]
        
        visited[i][j] = True
        
        for k in range(4):
            if self.isSafe(visited, grid, i + movex[k] , j + movey[k]):
                self.BFS(grid, i + movex[k], j + movey[k], visited)
```

```python
def count_shapes(A):
    N = len(A)
    M = len(A[0])
    
    ans = 0
    
    visited = [ [False]*(M) for _ in range(N)]

    for i in range(N):
        for j in range(M):
            if A[i][j] == 'X' and not visited[i][j]:
                  BFS(A, i, j, visited)
                  ans += 1
#                 q = []
#                 q.append((i, j))
#                 visited[i][j] = True
                
#                 while q:
#                     i, j = q.pop()
                    
#                     x = [0, 1, -1, 0]
#                     y = [-1, 0, 0, 1]
                    
#                     for dx, dy in zip(x, y):
#                         new_i = i + dx
#                         new_j = j + dy
#                         if is_safe(A, new_i, new_j, visited):
#                             visited[new_i][new_j] = True
#                             q.append((new_i, new_j))
                
#     print(visited)
    return ans
```
