#### Black Shapes

Given character matrix A of O's and X's, where O = white, X = black. Return the number of black shapes. A black shape consists of one or more adjacent X's (diagonals not included)

```python
def print_board(A):
    N = len(A)
    M = len(A[0])
    
    for i in range(N):
        for j in range(M):
            print(A[i][j], end = ' ')
        print()
```

```python
def is_safe(A, i, j, visited):
    N = len(A)
    M = len(A[0])
    
    if -1 < i < N and -1 < j < M and not visited[i][j] and A[i][j] == 'X':
        return True
    return False
```

```python
def BFS(grid, i, j, visited):
        
        movex = [-1, 0, 0, 1]
        movey = [ 0, -1, 1, 0]
        
        visited[i][j] = True
        
        for k in range(4):
            if is_safe(grid, i + movex[k] , j + movey[k],visited):
                BFS(grid, i + movex[k], j + movey[k], visited)
```

```python
def count_shapes(A):
    N = len(A)
    M = len(A[0])
    
    ans = 0
    
    visited = [ [False]*(M) for _ in range(N)]
    q = []
    for i in range(N):
        for j in range(len(A[0])):
#             print(i, j, A[i][j])
#             if i == 6 and j == 9:
#                 print(A[i][j], visited[i][j])
            if A[i][j] == 'X' and not visited[i][j]:
                #BFS(A, i, j, visited)
                ans += 1
                q.append((i, j))
                visited[i][j] = True
#                 A[i][j] = ans
                while q:
                    x, y = q.pop(0)
                    
                    dx = [0, 1, -1, 0]
                    dy = [-1, 0, 0, 1]
                    
                    for dx_x, dy_y in zip(dx, dy):
                        new_x = x + dx_x
                        new_y = y + dy_y
                        if is_safe(A, new_x, new_y, visited):
                            visited[new_x][new_y] = True
#                             A[new_i][new_j] = ans
                            q.append((new_x, new_y))
    
    
#     print_board(visited)
    return ans
```
